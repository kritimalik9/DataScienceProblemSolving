```python
import cv2
vid_capture = cv2.VideoCapture('video_sample.mp4') # Live video stream from web cam can also be taken as the input here
print(vid_capture)
while(vid_capture.isOpened()):
  # vid_capture.read() methods returns a tuple, first element is a bool 
  # and the second is frame
  ret, frame = vid_capture.read()
  if ret == True:
    cv2.imshow('Frame',frame)
    # 20 is in milliseconds, try to increase the value, say 50 and observe
    # key = cv2.waitKey(50)
    key = cv2.waitKey(50) & 0xFF
     
    if key == ord('q'): # 'q' ascii 113
      break
  else:
    break
vid_capture.release()
```

```python
# waitKey - already stopped the code if a key q pressed on the image. But need to destroy the window as well
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
```
