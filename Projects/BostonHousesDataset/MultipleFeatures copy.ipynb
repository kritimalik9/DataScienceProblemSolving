```python
def hypothesis(x, theta):
    yhat = 0.0
    n = x.shape[0] # x vector holding n features
    for i in range(n):
        yhat += theta[i] * x[i]
    return yhat
```

```python

def error(X, Y, theta):
    e = 0.0
    m = X.shape[0] # m examples
    for i in range(m):
        yhat = hypothesis(X[i], theta)
        e += (Y[i]-yhat)**2
    return e/m
```

```python

def gradient(X, Y, theta): # slope / derivative
    m,n = X.shape # 506 examples rows, 14 features cols
    # n gradients generated coz n features and thus n theta derivatives
    grad = np.zeros((n,))
    # theta = theta - learning rate * grad -> where both theta and grad are vectors of size n = 14
    for j in range(n):
        for i in range(m):
            yhat = hypothesis(X[i], theta)
            grad[j] += ((yhat - Y[i]) * X[i][j])
    return grad/m
```

```python
def gradient_descent(X, Y, learning_rate = 0.1, max_epochs = 300):
    m, n = X.shape
    theta = np.zeros((n,))
    error_list = []
    for i in range(max_epochs):
        # print("#############################################")
        # print(theta)
        e = error(X, Y, theta)
        error_list.append(e)
        grad = gradient(X, Y, theta)
        for j in range(n):
            theta[j] = theta[j] - learning_rate * grad[j]

    return theta, error_list
```

```python
import time
start = time.time()
theta, error_list = gradient_descent(X, Y)
end = time.time()
print(f"Time taken - {end - start}")
```

### Section-3 Optimizing Code using Vectorization
An efficient implementation for linear regression using vectorization
1. Avoid loops in the implementation - except gradient descent main loop
2. Use numpy functions like np.sum(), np.dot() which are quite fast and already optimized

```python
def hypothesis(X, theta):
    Yhatall = np.zeros(X.shape[1])
    Yhatall = np.dot(X, theta) # mxn matrix X multiply by nx1 size vector theta
    return Yhatall# prediction for all examples
```

```python
def error(X, Y, theta):
    Yhatall = np.dot(X, theta)
    total_error = np.sum((Yhatall - Y)**2) # np.sum over all 506 examples
    return total_error/X.shape[0] # div by m
```

```python
def gradient(X, Y, theta):
    grad = np.dot(X.T, (hypothesis(X,theta) - Y))
    m = X.shape[0]
    return grad / m
```

```python

def gradient_descent(X, Y, learning_rate=0.1, max_iters=300):
    theta = np.zeros((X.shape[1],)) # n features
    error_list = []
    for i in range(max_iters):
        error_list.append(error(X,Y, theta))
        grad = gradient(X, Y, theta)
        theta = theta - learning_rate * grad
    return theta, error_list
```

```python
start = time.time()
theta, error_list = gradient_descent(X, Y)
end = time.time()
print(f"Time taken by vectorized code = {end - start}")
```
