### Section 2 - Linear Regression

```python
# compute yhat = hypothesis
def hypothesis(x, theta):
    # theta = [theta0, theta1]
    return (theta[0] + (theta[1] * x))
```

```python
def gradient(X, Y, theta):
    grad = np.zeros((2,)) # creates a vector of two elements
    m = X.shape[0] # number of examples
    for i in range(m):
        grad[0] = grad[0] + (hypothesis(X[i], theta) - Y[i])
        grad[1] = grad[1] + (hypothesis(X[i], theta) - Y[i]) * X[i]
    return grad/m # divides both gradient by m
gradient(X, Y, [0, 0])
```

### Learning rate is a hyper parameter - try with different values to find out what works well

```python
def errorLossFn(X, Y, theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        total_error += (hypothesis(X[i], theta) - Y[i])** 2
    return total_error/m
```

```python
def gradientDescent(X, Y, max_steps = 100, learning_rate = 0.1):
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    for i in range(max_steps):
        grad = gradient(X, Y, theta)
        error = errorLossFn(X,Y,theta)
        print(grad)
        print(theta)
        error_list.append(error)
        theta_list.append((theta[0], theta[1]))
        print('*****')
        print('*****')
        theta0_stepsize = learning_rate * grad[0] # Mul grad by 2 generated from derivative not changing the final theta
        theta1_stepsize = learning_rate * grad[1] # Mul grad by 2 generated from derivative not changing the final theta
        theta[0] = theta[0] - theta0_stepsize
        theta[1] = theta[1] - theta1_stepsize
        # error_list.append(error)
        # theta_list.append((theta[0], theta[1]))
    return theta,error_list,theta_list
```

```python
theta, error_list, theta_list = gradientDescent(X, Y)
```

### Meaning of Accuracy - 97% ?
### Section - 4 Computing Score
### Score: R2 (R-Squared) or Coefficient of Determination

```python
def r_squared_score(Y, Yhat):
    num = np.sum((Y - Yhat)**2) # recommended over loops # variance(line)
    denom = np.sum((Y - Y.mean())**2) # variance(mean)
    score = (1 - (num/denom))
    return score*100
```

```python
r_squared_score(Y, Yhat)
```
