```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
X = pd.read_csv("./Training Data/Linear_X_Train.csv")
Y = pd.read_csv("./Training Data/Linear_Y_Train.csv")
theta = np.load("ThetaList.npy")
T0 = theta[:,0]
T1 = theta[:,1]
plt.ion()
for i in range(0, 50, 3): # plot after every 3 updates
    Yhat = T1[i]*X + T0[i]
    #points
    plt.scatter(X,Y)
    #line
    plt.plot(X,Yhat,'red')
    plt.draw() # draw the line
    plt.pause(1) # pause it for 1 sec
    plt.clf() # destroy the last object, to get an updated graph next time
```
