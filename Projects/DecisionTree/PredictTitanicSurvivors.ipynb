```python
def entropy(col):

    counts = np.unique(col, return_counts=True)
    N = float(col.shape[0])

    ent = 0.0

    for uniqueCategoryCnt in counts[1]:
        p = uniqueCategoryCnt/N
        ent += (-1.0 * p * np.log2(p))

    return ent
```

```python
X_test_df = pd.read_csv("./Test/Test.csv")
columns_to_drop = ['name', 'ticket', 'cabin', 'embarked', 'boat', 'body', 'home.dest']
X_test_cleaned = X_test_df.drop(columns_to_drop, axis=1)
X_test_cleaned['sex'] = le.transform(X_test_cleaned['sex'])
X_test_cleaned['age'].fillna(X_test_cleaned['age'].mean(), inplace=True)
print(X_test_cleaned.head(n=10))
print(X_test_cleaned.info())
```

```python
def divide_data(data_x_y, feature, feature_thres_val):
    left = pd.DataFrame([], columns=data_x_y.columns)
    right = pd.DataFrame([], columns=data_x_y.columns)
    num_samples = data_x_y.shape[0]

    for sample_idx in range(num_samples):
        if(data_x_y.loc[sample_idx][feature] > feature_thres_val):
            # right.append(data_x_y.loc[sample_idx])
            right.loc[len(right)] = data_x_y.loc[sample_idx]
        else:
            # left.append(data_x_y.loc[sample_idx])
            left.loc[len(left)] = data_x_y.loc[sample_idx]

    return left, right
```

```python
def informationGain(data_x_y, feature, feature_thres_val):
    num_samples = data_x_y.shape[0]

    l_data, r_data = divide_data(data_x_y, feature, feature_thres_val)

    l_data_sample_ratio = float(l_data.shape[0]/num_samples)
    r_data_sample_ratio = float(r_data.shape[0]/num_samples)

    if(l_data.shape[0] == 0 or r_data.shape[0] == 0):
        return -1000000

    info_gain = entropy(data_x_y.survived) - (l_data_sample_ratio * entropy(l_data.survived) + r_data_sample_ratio * entropy(r_data.survived))

    return info_gain
```

```python
# Test Information gain
for feature in X_train_cleaned.columns:
    print(feature)
    print(informationGain(X_train_cleaned, feature, X_train_cleaned[feature].mean()))
```

```python
class DecisionTree:
    def __init__(self, depth=0, maxdepth=5):
        self.left = None
        self.right = None
        self.depth = depth
        self.deciding_feature = None
        self.deciding_feature_thres = None
        self.Target = None
        self.max_depth = maxdepth

    def train(self, data_x_y):
        num_samples = data_x_y.shape[0]

        feature_list = [col for col in X_train_cleaned.columns if col != 'survived']

        info_gain_list = []

        for feature in feature_list:
            info_gain = informationGain(data_x_y, feature, data_x_y[feature].mean())
            info_gain_list.append(info_gain)
        
        max_info_gain_feature_idx = np.argmax(info_gain_list)
        max_info_gain_feature = feature_list[max_info_gain_feature_idx]

        self.deciding_feature = max_info_gain_feature
        self.deciding_feature_thres = data_x_y[max_info_gain_feature].mean()

        df_left, df_right = divide_data(data_x_y, self.deciding_feature, self.deciding_feature_thres)
        df_left = df_left.reset_index(drop=True)
        df_right = df_right.reset_index(drop=True)

        if(df_left.shape[0] == 0 or df_right.shape[0] == 0):
            if(data_x_y.survived.mean() >= 0.5):
                self.Target = "survived"
            else:
                self.Target = "not_survived"
            return
        if(self.depth > self.max_depth):
            if(data_x_y.survived.mean() >= 0.5):
                self.Target = "survived"
            else:
                self.Target = "not_survived"
            return
        self.left = DecisionTree(self.depth+1,self.max_depth)
        self.right = DecisionTree(self.depth+1,self.max_depth)
        self.left.train(df_left)
        self.right.train(df_right)
        if(data_x_y.survived.mean() >= 0.5):
            self.Target = "survived"
        else:
            self.Target = "not_survived"
        return

    def predict(self, testdata):
        if testdata[self.deciding_feature] > self.deciding_feature_thres:
            if self.right is None:
                return self.Target
            else:
                return self.right.predict(testdata)
        else:
            if self.left is None:
                return self.Target
            else:
                return self.left.predict(testdata)

        
```

```python
d = DecisionTree()
d.train(X_train_cleaned)
```

```python
pred = []
for sample_idx in range(len(X_test_cleaned)):
    predicted_class = d.predict(X_test_cleaned.loc[sample_idx])
    if(predicted_class == 'survived'):
        pred.append(1.0)
    else:
        pred.append(0.0)
```
