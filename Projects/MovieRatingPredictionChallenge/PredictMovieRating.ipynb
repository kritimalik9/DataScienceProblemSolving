```python
import pandas as pd

df = pd.read_csv("./Train.csv")
df.head(10)
```

```python
import re
def remove_tags(review_string):
    removelist = ""
    x = False
    # if x == False:
    #     x = True
    #     print(type(review_string))
    new_review_str = re.sub(r'http\S+', '', review_string) # remove links
    new_review_str = re.sub(r'<[^>]+>', '', new_review_str) # remove html tags
    new_review_str = re.sub(r'[^a-zA-Z]', ' ',new_review_str) # Remove punctuations and numbers
    new_review_str = re.sub(r"\s+[a-zA-Z]\s+", ' ', new_review_str) # Single character removal
    new_review_str = re.sub(r'\s+', ' ',new_review_str) # remove extra whitespaces
    return new_review_str.lower()

print(type(df['review']))
df_removed_tags = df.copy()
df_removed_tags['review'] = df['review'].apply(lambda r: remove_tags(r))
print(df.iloc[1, 0])
print(df_removed_tags.iloc[1, 0]) # file LN NUM 373
print(df.iloc[371, 0]) # file LN NUM 373
print(df_removed_tags.iloc[371, 0]) # file LN NUM 373
    
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python

stop_words = stopwords.words('english') # ex - this include both won and won't
df_removed_tags['review'] = df_removed_tags['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))

print(df_removed_tags.iloc[371, 0]) # file LN NUM 373
# df_removed_tags
```

```python
# Lemmatization

w_tokenize = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
def lemmatize_text(text):
    st = ''
    for w in w_tokenize.tokenize(text):
        st = st + lemmatizer.lemmatize(w, 'v') + ' '
    return st
df_lemmatized =  df_removed_tags
df_lemmatized['review'] = df_removed_tags['review'].apply(lemmatize_text)
print(df_removed_tags.iloc[371, 0]) # file LN NUM 373
print(df_lemmatized.iloc[371, 0]) # file LN NUM 373
```

```python
import math
def laplace_smoothing(label_word_count_dict, new_sent_word, new_sent_label, vocab):
    # print(f'label_word_count_dict[{new_sent_label}][{new_sent_word}] = {label_word_count_dict[new_sent_label][new_sent_word]}')
    numerator = label_word_count_dict[new_sent_label][new_sent_word] + 1
    num_words_given_label = sum(label_word_count_dict[new_sent_label].values())
    denominator = num_words_given_label + len(vocab)
    return math.log(numerator / float(denominator))
```

```python
def group_X_by_label(X, Y, unique_labels):
    X_unique_label_dict = {}
    for label in unique_labels:
        X_unique_label_dict[label] = X[np.where(Y == label)]
    return X_unique_label_dict
```

```python

def fit(X, Y, unique_labels):
    X_unique_label_dict = group_X_by_label(X, Y, unique_labels)
    num_examples = X.shape[0]
    log_label_priors = {}
    num_examples_given_label = {}
    for label, grouped_X in X_unique_label_dict.items():
        num_examples_given_label = grouped_X.shape[0]
        log_label_priors[label] = math.log(num_examples_given_label / (num_examples))# P(y=c)
    return log_label_priors
```

```python
Xtest_df = pd.read_csv('./Test/Test.csv')
print(Xtest_df.shape)
Xtest_df_removed_tags = Xtest_df.copy()
Xtest_df_removed_tags['review'] = Xtest_df['review'].apply(lambda r: remove_tags(r))
Xtest_df_removed_tags['review'] = Xtest_df_removed_tags['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
Xtest_df_lemmatized =  Xtest_df_removed_tags
Xtest_df_lemmatized['review'] = Xtest_df_removed_tags['review'].apply(lemmatize_text)
Xtest_reviews = Xtest_df_lemmatized['review'].values
# Xtest = vec.transform(Xtest_reviews) # transform using the same vocab of train data set
# Xtest_arr = Xtest.toarray()
# print(Xtest)
# print(Xtest_arr)
print(Xtest_reviews)
print(len(Xtest_reviews))
```

```python
def predict(vocab, label_word_count_dict, log_label_priors, unique_labels, Xtest_reviews):
    result = []
    for review in Xtest_reviews:
        # print(review)
        label_probs = {l : log_label_priors[l] for l in unique_labels}
        words = set(w_tokenize.tokenize(review))
        for word in words:
            if word not in vocab:
                continue
            reverse_index_dict = {w: index for index, w in enumerate(vocab)}
            for label in unique_labels:
                log_cond_prob_word_given_label = laplace_smoothing(label_word_count_dict, reverse_index_dict[word], label, vocab)
                label_probs[label] = label_probs[label] + log_cond_prob_word_given_label
        result.append(max(label_probs, key=label_probs.get))
        print(result[len(result) - 1])
        # break
    return result
        
        
```

```python
unique_labels = np.unique(encoded_labels)
log_label_priors = fit(X_arr, encoded_labels, unique_labels)
pred = predict(vocab, label_word_count_dict, log_label_priors, unique_labels, Xtest_reviews)
# pred
```
