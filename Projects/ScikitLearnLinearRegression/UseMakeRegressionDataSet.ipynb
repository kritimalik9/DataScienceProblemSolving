```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
# %matplotlib notebook
# %matplotlib widget
```

```python
# default make_regression - 100 examples with 100 features
X, Y = make_regression()
print(f'{X.shape}, {Y.shape}')
```

```python
X, Y = make_regression(n_samples=1000, n_features=2, n_informative=1, noise=10, random_state=1) # random_state = 1 means - get same data every time
print(f'{X.shape}, {Y.shape}')
```

```python
#  n_informative=1 - only 1 feature out of the given two features are used.
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], Y)
plt.show()

plt.subplot(1, 2, 2)
plt.scatter(X[:, 1], Y)
plt.show()
```

```python
X, Y = make_regression(n_samples=1000, n_features=2, n_informative=2, noise=10, random_state=1) # random_state = 1 means - get same data every time
print(f'{X.shape}, {Y.shape}')
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], Y)

plt.subplot(1, 2, 2)
plt.scatter(X[:, 1], Y)
plt.show()
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize= (10,7))
ax = plt.axes(projection = "3d")

ax.scatter3D(X[:,0], X[:,1], Y, color = 'green')
plt.title("3d scatter plot")

# Hyperplane eq / hypothesis = theta0 + theta1 * x1 + theta2 * x2
```

### IMPLEMENT Linear Regression Using Sk-learn

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, Y) # Train model -> preform Gradient Descent and find theta params
```

```python
model.coef_ # theta1, theta2, .. and so on
```

```python
model.intercept_ # theta0
```

```python
model.predict([X[0], X[1]]) # -> enter [] list of examples to predict
```
