```python
import cv2
import matplotlib.pyplot as plt
```

```python
bottom_left_img = cv2.imread('./ImageCollage dataset/bottom_left.jpg')
rdb_bottom_left_img = cv2.cvtColor(bottom_left_img, cv2.COLOR_BGR2RGB)
# print(rdb_bottom_left_img)
print(rdb_bottom_left_img.shape)
# plt.imshow(bottom_left_img)
plt.imshow(rdb_bottom_left_img)
crop_rdb_bottom_left_img = cv2.resize(rdb_bottom_left_img, (200,200))
plt.imshow(crop_rdb_bottom_left_img)
print(crop_rdb_bottom_left_img.shape)
```

```python
bottom_right_img = cv2.imread('./ImageCollage dataset/bottom_right.jpg')
rdb_bottom_right_img = cv2.cvtColor(bottom_right_img, cv2.COLOR_BGR2RGB)
# print(rdb_bottom_right_img)
print(rdb_bottom_right_img.shape)
# plt.imshow(bottom_left_img)
plt.imshow(rdb_bottom_right_img)
crop_rdb_bottom_right_img = cv2.resize(rdb_bottom_right_img, (200,200))
plt.imshow(crop_rdb_bottom_right_img)
print(crop_rdb_bottom_right_img.shape)
```

```python
center_img = cv2.imread('./ImageCollage dataset/center.jpeg')
rdb_center_img = cv2.cvtColor(center_img, cv2.COLOR_BGR2RGB)
# print(rdb_center_img)
print(rdb_center_img.shape)
# plt.imshow(bottom_left_img)
plt.imshow(rdb_center_img)
crop_rdb_center_img = cv2.resize(rdb_center_img, (100,100))
plt.imshow(crop_rdb_center_img)
print(crop_rdb_center_img.shape)
```

```python
top_left_img = cv2.imread('./ImageCollage dataset/top_left.jpg')
rdb_top_left_img = cv2.cvtColor(top_left_img, cv2.COLOR_BGR2RGB)
# print(rdb_top_left_img)
print(rdb_top_left_img.shape)
# plt.imshow(bottom_left_img)
plt.imshow(rdb_top_left_img)
crop_rdb_top_left_img = cv2.resize(rdb_top_left_img, (200,200))
plt.imshow(crop_rdb_top_left_img)
print(crop_rdb_top_left_img.shape)
```

```python
top_right_img = cv2.imread('./ImageCollage dataset/top_right.jpg')
rdb_top_right_img = cv2.cvtColor(top_right_img, cv2.COLOR_BGR2RGB)
# print(rdb_top_right_img)
print(rdb_top_right_img.shape)
# plt.imshow(bottom_left_img)
plt.imshow(rdb_top_right_img)
crop_rdb_top_right_img = cv2.resize(rdb_top_right_img, (200,200))
plt.imshow(crop_rdb_top_right_img)
print(crop_rdb_top_right_img.shape)
```

```python
import cv2
import matplotlib.pyplot as plt
bordered_bottom_left_img = cv2.copyMakeBorder(crop_rdb_bottom_left_img,0,10,10,10,borderType = cv2.BORDER_CONSTANT)
print(bordered_bottom_left_img.shape)
plt.imshow(bordered_bottom_left_img)
```

```python
import cv2
import matplotlib.pyplot as plt
bordered_bottom_right_img = cv2.copyMakeBorder(crop_rdb_bottom_right_img,0,10,0,10,borderType = cv2.BORDER_CONSTANT)
print(bordered_bottom_right_img.shape)
plt.imshow(bordered_bottom_right_img)
```

```python
import numpy as np
horz_bottom_img = np.hstack([bordered_bottom_left_img, bordered_bottom_right_img])
plt.imshow(horz_bottom_img)
print(horz_bottom_img.shape)
```

```python

bordered_top_left_img = cv2.copyMakeBorder(crop_rdb_top_left_img,10,10,10,10,borderType = cv2.BORDER_CONSTANT)
print(bordered_top_left_img.shape)
plt.imshow(bordered_top_left_img)
```

```python
bordered_top_right_img = cv2.copyMakeBorder(crop_rdb_top_right_img,10,10,0,10,borderType = cv2.BORDER_CONSTANT)
print(bordered_top_right_img.shape)
plt.imshow(bordered_top_right_img)
```

```python

horz_top_img = np.hstack([bordered_top_left_img, bordered_top_right_img])
plt.imshow(horz_top_img)
print(horz_top_img.shape)
```

```python
result_img_temp = np.vstack([horz_top_img, horz_bottom_img])
plt.imshow(result_img_temp)
print(result_img_temp.shape)
```

```python
import cv2
import matplotlib.pyplot as plt
bordered_center_img = cv2.copyMakeBorder(crop_rdb_center_img,10,10,10,10,borderType = cv2.BORDER_CONSTANT)
print(bordered_center_img.shape)
plt.imshow(bordered_center_img)
```

```python
result_img_temp[155 : 275, 155 : 275, :] = bordered_center_img
print(result_img_temp)
plt.imshow(result_img_temp)
print(result_img_temp.shape)
```

```python
import pandas as pd
np_array = np.array(result_img_temp)
flatten_arr = np_array.reshape(430*430, 3)
print(flatten_arr.shape)
df = pd.DataFrame(flatten_arr)
df.columns = ['r', 'g', 'b']
df.to_csv('result.csv', index = False)
# df = 
```
